There is no null values but we will do to fix because its for learning and helful for that cases if we face any missing values problems this data set is from kaggle.

```python
df = df.dropna()
```

```python
df["experience_years"] = df["experience_years"].fillna(df["experience_years"].median())
```

Replace with median


```python
df["experience_years"] = df["experience_years"].fillna(df["experience_years"].mean())
```

replace with mean

```python
df["education_level"] = df["education_level"].fillna(df["education_level"].mode()[0])
```

Replace with most frequent value

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

Remove duplicates if any

```python
df.dtypes
```

check data types if any data type is incorrect so we will fix it first

```python
df["salary"] = df["salary"].astype(float)
```

```python
df["education_level"].unique()
```

```python
df["education_level"] = df["education_level"].replace({
    "bachelor":"Bachelor",
    "Bachelors":"Bachelor",
    "Bachelor Degree":"Bachelor",
    "masters":"Master",
    "Master Degree":"Master",
    "phd":"PhD"
})
```

if you see some mistakes so you can check it first

```python
df["job_title"] = df["job_title"].replace({
    "Software Eng.":"Software Engineer",
    "Software Developer":"Software Engineer",
    "SWE":"Software Engineer"
})
```

```python
df["job_title"] = df["job_title"].str.strip()
```

```python
text_columns = df.select_dtypes(include="object").columns

for col in text_columns:
    df[col] = df[col].str.strip()
```

remove All extra spaces in text coloumns

```python
df["education_level"] = df["education_level"].str.title()
```

```python
upper = Q3 + 1.5*IQR
```
